```python
import re
from urllib.request import urlopen
import urllib
from bs4 import BeautifulSoup
import csv
import pandas as pd
import time
import datetime
import numpy as np
import os
import string
import json
from http import cookiejar
```

```python
def format_filename(s):
#Take a string and return a valid filename constructed from the string. 
#Uses a whitelist approach: any characters not present in valid_chars are removed. 
    valid_chars = "-_.() %s%s" % (string.ascii_letters, string.digits)
    filename = ''.join(c for c in s if c in valid_chars)
    return filename
```

```python
title=[]
link=[]
date=[] 
begin='20150101'
end='20151231'
query='democracy'
api_key='378cb0f502d779d371bf3257c93a616c:17:74686526'
for page in range(101):
    try:
        url = 'http://api.nytimes.com/svc/search/v2/articlesearch.json?q='+ query + '&fq=type_of_material%3A%28%22News%22%29+AND+source%3A%28%22The+New+York+Times%22%29'+'&begin_date=' + begin + '&end_date=' + end +'&sort=newest' +'&page=' + str(page) + '&api-key=' + api_key
        str_content = urlopen(url).read().decode('utf-8')
        articles = json.loads(str_content)
        for article in articles["response"]["docs"]:
            tit= article["headline"]["main"].replace("\n","")
            lin= article["web_url"]
            dat = article["pub_date"] 
            title.append(tit) 
            link.append(lin)       
            date.append(dat)
    except:
        continue
```

```python
#need to enable cookie, or the page will redirect it self into a infinite loop.
cj = cookiejar.CookieJar()
# Create opener
opener = urllib.request.build_opener(urllib.request.HTTPCookieProcessor(cj))
for linkk,datee,titlee in zip(link,date,title):
    time.sleep(1)
    dt=' '.join([datee, titlee])
    name = '{}.txt'.format(dt)
    filename=format_filename(name)
    html = opener.open(linkk).read()
    soup = BeautifulSoup(html,'lxml')
    out_dir='nytimes'
    os.makedirs(out_dir,exist_ok=True)
    with open(out_dir + "/" + filename, 'w',encoding='utf-8') as f:
        for body in soup.find_all('p',attrs={'itemprop':'articleBody'}):
            f.write(body.get_text())
```
